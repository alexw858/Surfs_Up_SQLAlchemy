```python
import pandas as pd
import numpy as np
import datetime as dt
import matplotlib.pyplot as plt

import sqlalchemy
from sqlalchemy.ext.automap import automap_base
from sqlalchemy.orm import Session
from sqlalchemy import create_engine, inspect, func

from collections import Counter
```

```python
engine = create_engine('sqlite:///./Resources/hawaii.sqlite')
```

```python
Base = automap_base()

Base.prepare(engine, reflect=True)

Measurement = Base.classes.measurements
Station = Base.classes.stations

session = Session(engine)
```

```python
inspector = inspect(engine)
inspector.get_table_names()
```

```python
# Design a query to retrieve the last 12 months of precipitation data.
m_results=session.query(Measurement.date, Measurement.prcp, Measurement.tobs, Measurement.station).all()
precipitation = [r[1] for r in m_results]
print(precipitation)
```

```python
#Select only the `date` and `prcp` values
date = [r[0] for r in m_results]
print(date, precipitation)
```

```python
df = pd.DataFrame(m_results, columns=['date', 'precipitation', 'tobs', 'station']).set_index('date')
df.head()
```

```python
# Plot the results using the DataFrame `plot` method.
plt.title("Precipitation Analysis")
# x_axis = np.arange(0, 14)
plt.xticks(np.arange(len(df), step=120), rotation='vertical')
plt.xlabel("date")
plt.ylabel("precipitation")
plt.plot(df.reset_index()["date"], df["precipitation"])
plt.show()
```

```python
s_results = session.query(Station.station, Station.name, Station.latitude, \
                         Station.longitude, Station.elevation)
stations = [s[0] for s in s_results]
len(stations)
```

```python
# Design a query to find the most active stations.
df.groupby(["station"]).count().sort_values(by='tobs', ascending=False).rename(columns={'tobs': 'observation count'})\
.drop(['precipitation'], axis=1)
```

```python
# Design a query to retrieve the last 12 months of temperature observation data (tobs).
tobs = [t[2] for t in m_results]
tobs
```

```python
#Filter by the station with the highest number of observations.
tobs_results = engine.execute('select tobs from measurements where station="USC00519281"').fetchall()
tobs_results
# f = [tuple(tobs_results)]
```

```python
plt.hist(tobs_results, 12)
plt.xlabel("Temperature")
plt.ylabel("Frequency")
plt.title("Temp Histogram")
plt.grid(True)
plt.show()
```
